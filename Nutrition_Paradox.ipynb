```python
# Define a dictionary of URLs
urls = {
    'obesity_adults': "https://ghoapi.azureedge.net/api/NCD_BMI_30C",
    'obesity_children': "https://ghoapi.azureedge.net/api/NCD_BMI_PLUS2C",
    'malnutrition_adults': "https://ghoapi.azureedge.net/api/NCD_BMI_18C",
    'malnutrition_children': "https://ghoapi.azureedge.net/api/NCD_BMI_MINUS2C"
}

# Fetch data and create DataFrames
df_obesity_adults = pd.DataFrame(requests.get(urls['obesity_adults']).json()['value'])
df_obesity_adults['age_group'] = 'adults'

df_obesity_children = pd.DataFrame(requests.get(urls['obesity_children']).json()['value'])
df_obesity_children['age_group'] = 'children'

df_malnutrition_adults = pd.DataFrame(requests.get(urls['malnutrition_adults']).json()['value'])
df_malnutrition_adults['age_group'] = 'adults'

df_malnutrition_children = pd.DataFrame(requests.get(urls['malnutrition_children']).json()['value'])
df_malnutrition_children['age_group'] = 'children'

# Display the head of each DataFrame to verify
print("Obesity Adults:")
display(df_obesity_adults.head())
print("\nObesity Children:")
display(df_obesity_children.head())
print("\nMalnutrition Adults:")
display(df_malnutrition_adults.head())
print("\nMalnutrition Children:")
display(df_malnutrition_children.head())
```

```python
# Combine obesity datasets
df_obesity = pd.concat([df_obesity_adults, df_obesity_children], ignore_index=True)

# Combine malnutrition datasets
df_malnutrition = pd.concat([df_malnutrition_adults, df_malnutrition_children], ignore_index=True)

# Convert 'TimeDim' to int for filtering
df_obesity['TimeDim'] = df_obesity['TimeDim'].astype(int)
df_malnutrition['TimeDim'] = df_malnutrition['TimeDim'].astype(int)

# Filter data between 2012 and 2022 inclusive
df_obesity = df_obesity[(df_obesity['TimeDim'] >= 2012) & (df_obesity['TimeDim'] <= 2022)]
df_malnutrition = df_malnutrition[(df_malnutrition['TimeDim'] >= 2012) & (df_malnutrition['TimeDim'] <= 2022)]

# Quick check
print("Filtered Obesity Data:")
display(df_obesity.head())

print("Filtered Malnutrition Data:")
display(df_malnutrition.head())
```

```python
# Define the columns to keep
columns_to_keep = ['ParentLocation', 'Dim1', 'TimeDim', 'Low', 'High', 'NumericValue', 'SpatialDim', 'age_group']

# Keep only the selected columns in both datasets
df_obesity = df_obesity[columns_to_keep]
df_malnutrition = df_malnutrition[columns_to_keep]

# Display heads for verification
print("Cleaned Obesity Dataset:")
print(df_obesity.head())

print("\nCleaned Malnutrition Dataset:")
print(df_malnutrition.head())
```

```python
df_obesity.rename(columns={
    'TimeDim': 'Year',
    'Dim1': 'Gender',
    'NumericValue': 'Mean_Estimate',
    'Low': 'LowerBound',
    'High': 'UpperBound',
    'ParentLocation': 'Region',
    'SpatialDim': 'Country'
}, inplace=True)

df_malnutrition.rename(columns={
    'TimeDim': 'Year',
    'Dim1': 'Gender',
    'NumericValue': 'Mean_Estimate',
    'Low': 'LowerBound',
    'High': 'UpperBound',
    'ParentLocation': 'Region',
    'SpatialDim': 'Country'
}, inplace=True)
```

```python
# Define gender mapping
gender_mapping = {
    'SEX_MLE': 'Male',
    'SEX_FMLE': 'Female',
    'SEX_BTSX': 'Both'
}

# Apply mapping to both datasets
df_obesity['Gender'] = df_obesity['Gender'].map(gender_mapping)
df_malnutrition['Gender'] = df_malnutrition['Gender'].map(gender_mapping)
```

```python
# Function to convert ISO Alpha-3 codes to full country names
def get_country_name(code):
    try:
        return pycountry.countries.get(alpha_3=code).name
    except:
        # Return None if code is not found
        return None

# Dictionary for special WHO/World Bank codes
special_cases = {
    'GLOBAL': 'Global',
    'WB_LMI': 'Low & Middle Income',
    'WB_HI': 'High Income',
    'WB_LI': 'Low Income',
    'EMR': 'Eastern Mediterranean Region',
    'EUR': 'Europe',
    'AFR': 'Africa',
    'SEAR': 'South-East Asia Region',
    'WPR': 'Western Pacific Region',
    'AMR': 'Americas Region',
    'WB_UMI': 'Upper Middle Income'
}

# Apply function to convert country codes
def convert_country_code(code):
    name = get_country_name(code)
    if name:
        return name
    else:
        return special_cases.get(code, code)  # default to original if not found

# Apply the function to convert country codes in both dataframes
df_obesity['Country'] = df_obesity['Country'].apply(convert_country_code)
df_malnutrition['Country'] = df_malnutrition['Country'].apply(convert_country_code)
```

## Summary of Findings

Based on the visualizations of obesity and malnutrition rates from 2012 to 2022:

### Obesity Trends

*   **Overall Trend:** Obesity rates generally show an increasing trend over the years for both adults and children.
*   **Age Group Comparison:** Adults consistently have significantly higher obesity rates compared to children. The gap between adult and child obesity rates appears to be widening over time.
*   **Gender Comparison:** For adults, the obesity rates for males and females are relatively similar and follow a similar increasing trend. For children, the obesity rates for both genders are also similar, but at a much lower level than adults.
*   **Regional Comparison:** The mean obesity rates vary by region. The Western Pacific and Americas regions appear to have higher mean obesity rates compared to other regions like Africa and South-East Asia.

### Malnutrition Trends

*   **Overall Trend:** Malnutrition rates show a general decreasing trend over the years for both adults and children, although the changes are less pronounced than the increase in obesity.
*   **Age Group Comparison:** Malnutrition rates are generally higher in children compared to adults, although the difference is not as large as the difference in obesity rates between age groups.
*   **Gender Comparison:** Malnutrition rates are relatively similar for males and females across both age groups.
*   **Regional Comparison:** Malnutrition rates also vary by region. South-East Asia and Africa appear to have higher mean malnutrition rates compared to other regions like Europe and the Americas.

### Obesity vs Malnutrition Comparison

*   Obesity rates are generally higher than malnutrition rates across all age groups, genders, and regions in the 2012-2022 period.
*   The gap between obesity and malnutrition rates is particularly large for adults.
*   While obesity rates are increasing and malnutrition rates are decreasing, both remain significant public health concerns in different populations and regions.

```python
conn = sqlite3.connect('nutrition_analysis.db')
cursor = conn.cursor()

# Create tables
cursor.execute('''
CREATE TABLE IF NOT EXISTS obesity (
    Year INTEGER,
    Gender TEXT,
    Mean_Estimate REAL,
    LowerBound REAL,
    UpperBound REAL,
    age_group TEXT,
    Country TEXT,
    Region TEXT,
    CI_Width REAL,
    Obesity_Level TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS malnutrition (
    Year INTEGER,
    Gender TEXT,
    Mean_Estimate REAL,
    LowerBound REAL,
    UpperBound REAL,
    age_group TEXT,
    Country TEXT,
    Region TEXT,
    CI_Width REAL,
    Malnutrition_Level TEXT
)
''')

# Calculate CI_Width and add placeholder columns
df_obesity['CI_Width'] = df_obesity['UpperBound'] - df_obesity['LowerBound']
df_obesity['Obesity_Level'] = None  # Placeholder

df_malnutrition['CI_Width'] = df_malnutrition['UpperBound'] - df_malnutrition['LowerBound']
df_malnutrition['Malnutrition_Level'] = None # Placeholder


# Insert into obesity
for _, row in df_obesity.iterrows():
    cursor.execute('''
    INSERT INTO obesity VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', (
        row['Year'], row['Gender'], row['Mean_Estimate'], row['LowerBound'], row['UpperBound'],
        row['age_group'], row['Country'], row['Region'], row['CI_Width'], row['Obesity_Level']
    ))

# Insert into malnutrition
for _, row in df_malnutrition.iterrows():
    cursor.execute('''
    INSERT INTO malnutrition VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', (
        row['Year'], row['Gender'], row['Mean_Estimate'], row['LowerBound'], row['UpperBound'],
        row['age_group'], row['Country'], row['Region'], row['CI_Width'], row['Malnutrition_Level']
    ))

conn.commit()
conn.close()
```

```python
conn = sqlite3.connect('nutrition_analysis.db')
cursor = conn.cursor()

# 1. Top 5 regions with highest avg obesity in 2022
print("1. Top 5 regions with highest avg obesity in 2022:")
cursor.execute('''
SELECT Region, AVG(Mean_Estimate) AS Avg_Obesity
FROM obesity
WHERE Year = 2022
GROUP BY Region
ORDER BY Avg_Obesity DESC
LIMIT 5;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)

# 2. Top 5 countries with highest obesity estimates
print("2. Top 5 countries with highest obesity estimates:")
cursor.execute('''
SELECT Country, MAX(Mean_Estimate) AS Max_Obesity
FROM obesity
GROUP BY Country
ORDER BY Max_Obesity DESC
LIMIT 5;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)

# 3. Obesity trend in India
print("3. Obesity trend in India:")
cursor.execute('''
SELECT Year, AVG(Mean_Estimate) AS Avg_Obesity
FROM obesity
WHERE Country = 'India'
GROUP BY Year
ORDER BY Year;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)


# 4. Average obesity by gender
print("4. Average obesity by gender:")
cursor.execute('''
SELECT Gender, AVG(Mean_Estimate) AS Avg_Obesity
FROM obesity
GROUP BY Gender;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)

# 5. Country count by obesity level & age group - Note: Obesity_Level is None in current data
print("5. Country count by obesity level & age group:")
cursor.execute('''
SELECT Obesity_Level, age_group, COUNT(DISTINCT Country) AS Country_Count
FROM obesity
GROUP BY Obesity_Level, age_group;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)

# 6. Top 5 least reliable & top 5 most consistent countries (CI_Width)
# Least reliable
print("6. Top 5 least reliable countries (highest Avg_CI):")
cursor.execute('''
SELECT Country, AVG(CI_Width) AS Avg_CI
FROM obesity
GROUP BY Country
ORDER BY Avg_CI DESC
LIMIT 5;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)

# Most consistent
print("Top 5 most consistent countries (lowest Avg_CI):")
cursor.execute('''
SELECT Country, AVG(CI_Width) AS Avg_CI
FROM obesity
GROUP BY Country
ORDER BY Avg_CI ASC
LIMIT 5;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)

# 7. Average obesity by age group
print("7. Average obesity by age group:")
cursor.execute('''
SELECT age_group, AVG(Mean_Estimate) AS Avg_Obesity
FROM obesity
GROUP BY age_group;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)

# 8. Top 10 countries with consistently low obesity
print("8. Top 10 countries with consistently low obesity:")
cursor.execute('''
SELECT Country, AVG(Mean_Estimate) AS Avg_Obesity, AVG(CI_Width) AS Avg_CI
FROM obesity
GROUP BY Country
HAVING Avg_Obesity < 25 AND Avg_CI < 3
ORDER BY Avg_Obesity ASC
LIMIT 10;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)

# 9. Countries where female obesity > male obesity by large margin (same year)
print("9. Countries where female obesity > male obesity by large margin (same year):")
cursor.execute('''
SELECT o1.Country, o1.Year, o1.Mean_Estimate AS Female_Obesity, o2.Mean_Estimate AS Male_Obesity,
       (o1.Mean_Estimate - o2.Mean_Estimate) AS Difference
FROM obesity o1
JOIN obesity o2 ON o1.Country = o2.Country AND o1.Year = o2.Year
WHERE o1.Gender = 'Female' AND o2.Gender = 'Male'
AND (o1.Mean_Estimate - o2.Mean_Estimate) > 5;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)

# 10. Global average obesity per year
print("10. Global average obesity per year:")
cursor.execute('''
SELECT Year, AVG(Mean_Estimate) AS Global_Obesity_Avg
FROM obesity
GROUP BY Year
ORDER BY Year;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)

conn.close()
```

```python
conn = sqlite3.connect('nutrition_analysis.db')
cursor = conn.cursor()

# 1. Obesity vs Malnutrition comparison by country (example: top 5)
print("1. Obesity vs Malnutrition comparison by country (example: top 5):")
cursor.execute('''
SELECT o.Country,
       AVG(o.Mean_Estimate) AS Avg_Obesity,
       AVG(m.Mean_Estimate) AS Avg_Malnutrition
FROM obesity o
JOIN malnutrition m ON o.Country = m.Country AND o.Year = m.Year
GROUP BY o.Country
ORDER BY Avg_Obesity DESC
LIMIT 5;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)

# 2. Gender-based disparity in both
print("2. Gender-based disparity in both:")
cursor.execute('''
SELECT o.Gender,
       AVG(o.Mean_Estimate) AS Obesity,
       AVG(m.Mean_Estimate) AS Malnutrition
FROM obesity o
JOIN malnutrition m ON o.Gender = m.Gender AND o.Year = m.Year
GROUP BY o.Gender;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)

# 3. Region-wise avg (Africa and America)
print("3. Region-wise avg (Africa and America):")
cursor.execute('''
SELECT o.Region,
       AVG(o.Mean_Estimate) AS Avg_Obesity,
       AVG(m.Mean_Estimate) AS Avg_Malnutrition
FROM obesity o
JOIN malnutrition m ON o.Region = m.Region AND o.Year = m.Year
WHERE o.Region IN ('Africa', 'Americas Region')
GROUP BY o.Region;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)

# 4. Countries with rising obesity + falling malnutrition
print("4. Countries with rising obesity + falling malnutrition:")
cursor.execute('''
SELECT o.Country,
       MAX(o.Mean_Estimate) - MIN(o.Mean_Estimate) AS Obesity_Increase,
       MIN(m.Mean_Estimate) - MAX(m.Mean_Estimate) AS Malnutrition_Decrease
FROM obesity o
JOIN malnutrition m ON o.Country = m.Country AND o.Year = m.Year
GROUP BY o.Country
HAVING Obesity_Increase > 5 AND Malnutrition_Decrease < 0;
''')

# 5. Age-wise trend analysis
print("5. Age-wise trend analysis:")
cursor.execute('''
SELECT o.age_group,
       AVG(o.Mean_Estimate) AS Avg_Obesity,
       AVG(m.Mean_Estimate) AS Avg_Malnutrition
FROM obesity o
JOIN malnutrition m
    ON o.age_group = m.age_group
    AND o.Year = m.Year
GROUP BY o.age_group
ORDER BY o.age_group;
''')
results = cursor.fetchall()
for row in results:
    print(row)
print("-" * 20)
```

```python
import pandas as pd
import sqlite3
import plotly.express as px
import streamlit as st

# Title
st.title("🌍 Global Obesity & Malnutrition Dashboard")

# Connect to DB
conn = sqlite3.connect('nutrition_analysis.db')

# Load data
df_obesity = pd.read_sql_query("SELECT * FROM obesity", conn)
df_malnutrition = pd.read_sql_query("SELECT * FROM malnutrition", conn)

# Merge for joint analysis
df_combined = pd.merge(df_obesity, df_malnutrition,
                       on=['Country', 'Year', 'Gender', 'age_group'],
                       suffixes=('_obesity', '_malnutrition'))

# Sidebar Filters
st.sidebar.header("Filter Data")
selected_country = st.sidebar.selectbox("Select Country", sorted(df_combined['Country'].unique()))
selected_year = st.sidebar.slider("Select Year",
                                   int(df_combined['Year'].min()),
                                   int(df_combined['Year'].max()),
                                   step=1)

# Filtered Data
filtered_data = df_combined[(df_combined['Country'] == selected_country) &
                            (df_combined['Year'] == selected_year)]

# Section 1: Country Overview
st.subheader(f"📌 {selected_country} - Year {selected_year}")

col1, col2 = st.columns(2)

with col1:
    st.metric("Avg Obesity (%)", round(filtered_data['Mean_Estimate_obesity'].mean(), 2))

with col2:
    st.metric("Avg Malnutrition (%)", round(filtered_data['Mean_Estimate_malnutrition'].mean(), 2))

# Section 2: Gender Comparison
st.subheader("🔍 Gender Comparison")
gender_plot = px.bar(filtered_data, x='Gender',
                     y=['Mean_Estimate_obesity', 'Mean_Estimate_malnutrition'],
                     barmode='group',
                     labels={"value": "Estimate (%)", "variable": "Indicator"},
                     title="Obesity vs Malnutrition by Gender")
st.plotly_chart(gender_plot)

# Section 3: Age-wise Comparison
st.subheader("📊 Age-wise Trend")
age_plot = px.line(filtered_data, x='age_group',
                   y=['Mean_Estimate_obesity', 'Mean_Estimate_malnutrition'],
                   labels={"value": "Estimate (%)", "variable": "Indicator"},
                   title="Obesity & Malnutrition by Age Group")
st.plotly_chart(age_plot)

# Section 4: Global Scatter
st.subheader("🌐 Global Obesity vs Malnutrition (All Countries)")

avg_global = df_combined.groupby("Country")[['Mean_Estimate_obesity', 'Mean_Estimate_malnutrition']].mean().reset_index()

scatter = px.scatter(avg_global, x='Mean_Estimate_obesity', y='Mean_Estimate_malnutrition',
                     hover_name='Country',
                     labels={"Mean_Estimate_obesity": "Obesity (%)", "Mean_Estimate_malnutrition": "Malnutrition (%)"},
                     title="Obesity vs Malnutrition - Country-wise Comparison")
st.plotly_chart(scatter)

# Countries where Obesity Increased and Malnutrition Decreased
st.subheader("Countries where Obesity Increased and Malnutrition Decreased")

change_df = df_combined.groupby('Country').agg(
    obesity_min=('Mean_Estimate_obesity', 'min'),
    obesity_max=('Mean_Estimate_obesity', 'max'),
    malnutrition_min=('Mean_Estimate_malnutrition', 'min'),
    malnutrition_max=('Mean_Estimate_malnutrition', 'max')
).reset_index()

change_df['obesity_diff'] = change_df['obesity_max'] - change_df['obesity_min']
change_df['malnutrition_diff'] = change_df['malnutrition_max'] - change_df['malnutrition_min']

trend_df = change_df[
    (change_df['obesity_diff'] > 0) &
    (change_df['malnutrition_diff'] < 0)
].sort_values(by='obesity_diff', ascending=False)

st.table(trend_df[['Country', 'obesity_diff', 'malnutrition_diff']].head(10))

# Region-wise Comparison (Obesity vs Malnutrition)
st.subheader("Region-wise Obesity vs Malnutrition Averages")

region_avg = filtered_data.groupby('Region_obesity')[['Mean_Estimate_obesity', 'Mean_Estimate_malnutrition']].mean().reset_index()

fig_region = px.bar(
    region_avg, x='Region_obesity', y=['Mean_Estimate_obesity', 'Mean_Estimate_malnutrition'],
    barmode='group',
    labels={'value': 'Average (%)', 'Region_obesity': 'Region', 'variable': 'Indicator'},
    title='Average Obesity vs Malnutrition by Region'
)
st.plotly_chart(fig_region, use_container_width=True)


# Close DB connection
conn.close()
```
